## List comprehension in Python

list comprehension is a consise way of creating a new list from an existing one without the use of a for loop

syntax:




new_list = [expression for items in iterable]
    Where:

*   expression: The value to add to the new list
*   item: The current value from the iterable.
* iterable: The collection you are iterating over.

### examples

```python
# To create a list of Numbers from 1 to 10

numbers = [n for n in range(0, 11)]

# Squaring Numbers
squares = [n * n for n in range(10)]

# Finding length of each word in a list
words = ["hello", "world", "python"]
lengths = [len(word) for word in words]
```

# Compound datatype in Python

Compound datatypes are types of data made up of two or more values or elements
Examples of such data types are lists, tuples and sets

## Lists
Lists are data structures used to store a collection of items.

Lists are mutable, meaning you can add items, remove or modify once created.

Example

```python
# Example creating a list
my_list =['Ruth', 19, 'mustard']
print(my_list[0])
```

## Tuples
A tuple is an ordered collection of elements which can be of different data types

Tuples are immutable  i.e the elements cannot be changed once defined

```python
# Creating a tuple
tuple = (2.4, "gauge", "True")
print(tuple[2])
```

## Sets
Sets are a collection of unique elements i.e there are no duplicates
Like lists, sets are mutable.
However, sets are unordered, unlike lists, which are ordered collections. This means the elements in a set do not maintain any particular order, whereas lists do.

```python
# Creating a set
fruits ={"melon", "mangoes", "oranges"}
print(fruits)
```
